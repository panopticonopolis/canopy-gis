```python
import ee
import folium
import geemap.eefolium as gmap
try:
    ee.Initialize()
except Exception as e:
    ee.Authenticate()
    ee.Initialize() 

TCI_RGB = ['b1', 'b2', 'b3']
vis = {'bands': TCI_RGB,"max": 3000, "gamma": [3, 3.1, 2.9]}
```

```python
def visualize_col(col_path):
    Map = gmap.Map()
    basemap_col = ee.ImageCollection(col_path)
    mosaic = basemap_col.mosaic()
    Map.setCenter(19.073072,0.156041,5)
    Map.addLayer(mosaic, vis)
    Map.add_child(folium.LayerControl())
    return Map

visualize_col("users/zwarshavsky/basemap_v1")

```
