# Duplicate tile analysis

```python
#Additional Params
bands = ['R10m/TCI']

#Republic of the Congo Bounding Box
search_bbox = BBox(bbox=[11.6395900000061374,-4.8205900000029942, 18.4909000000043306,3.6516899999987800], crs=CRS.WGS84)


def detect_duplicates(start_years):
    for result in start_years:
        search_time_interval = (f'{result}-01-01T00:00:00', '2020-12-31T23:59:59')
        wfs_iterator = WebFeatureService(
            search_bbox,
            search_time_interval,
            data_source=DataSource.SENTINEL2_L2A,
            maxcc=.05,
            config=config
        )
        results = wfs_iterator.get_tiles()
        df = pd.DataFrame(results, columns=['Tilename','Date','AmazonID'])
        print("Between",result, "and August 2020")
        print("Total tiles:",len(df))
        print("Total unique tiles:",len(df.drop_duplicates(subset='Tilename')))
        print("% Tiles < 10 duplicate:", (df["Tilename"].value_counts() < 10).value_counts()[1] / len(df.drop_duplicates(subset='Tilename')))
        print("---------------------")
        print("\n")
```

# Load CSV - New

```python
import json

df_labels = pd.read_csv("./data/labelled/labels_Misha_v2.csv")
df_labels = df_labels[["center-lat","center-long","polygon"]][0:33]
```

```python
df_labels
```

```python
polygons = []
for polygon in df_labels["polygon"]:
    polygons.append(json.loads(polygon)["coordinates"])
```

```python
coordinates = []
for items in polygons:
    for item in items:
        for lon_lat in item:
            coordinates.append(lon_lat)
```

```python
#bounding box

min_lon = min([i[0] for i in coordinates])
min_lat = min([i[1] for i in coordinates])
max_lon = max([i[0] for i in coordinates])
max_lat = max([i[1] for i in coordinates])
```

```python
bounding_box = min_lon,min_lat,max_lon,max_lat
```

```python
bounding_box
```

```python
tiles = []
for ll in coordinates:
    tiles.append(encode(LLtoUTM(ll[1],ll[0]),1)[:-2])

tiles = list(set(tiles))
```

# Downloading Data in Both Formats

```python
#Congo Basin Bounding Box
# search_bbox = BBox(bbox=[6.81,-5.27, 30, 5.5], crs=CRS.WGS84)

#Additional Params
bands = ['R10m/TCI']

#Misha's Tiles of Interest
search_bbox = BBox(bbox=bounding_box, crs=CRS.WGS84)



search_time_interval = ('2019-01-01T00:00:00', '2020-12-31T23:59:59')
wfs_iterator = WebFeatureService(
    search_bbox,
    search_time_interval,
    data_source=DataSource.SENTINEL2_L2A,
    maxcc=.1,
    config=config
)
results = wfs_iterator.get_tiles()
df = pd.DataFrame(results, columns=['Tilename','Date','AmazonID'])
df_tiles_of_interest = df[df["Tilename"].isin(tiles)]
df2 = (df_tiles_of_interest.groupby('Tilename').head(10))
output2 = list(df2.itertuples(index=False,name=None))
for tile in output2:
    tile_name, time, aws_index = tile
    data_folder_safe = '/Volumes/Lacie/zhenyadata/Project_Canopy_Data/PC_Data/Sentinel_Data/Labelled/Tiles_v2_Misha/SAFE_true_raw'
    data_folder_safe_false = '/Volumes/Lacie/zhenyadata/Project_Canopy_Data/PC_Data/Sentinel_Data/Labelled/Tiles_v2_Misha/SAFE_false_raw'
    
    #Download SAFE Files
    request = AwsTileRequest(
        tile=tile_name,
        time=time,
        bands = bands, 
        aws_index=aws_index,
        data_folder=data_folder_safe,
        data_source=DataSource.SENTINEL2_L2A,
        safe_format = True
    )

    request.save_data(redownload=True)
    
    #Download SAFE false Files
    
    request = AwsTileRequest(
    tile=tile_name,
    time=time,
    aws_index=aws_index,
    data_folder=data_folder_safe_false,
    data_source=DataSource.SENTINEL2_L2A,
    safe_format = False
    )

    request.save_data(redownload=True)
```
