### MGRS Test

The goal of this notebook is to get a list of all the MGRS Tile IDs in the congo basin.
It uses the following method:

1. Draw a bounding box across the entire congo basin rainforest region
2. Make a list of evenly-spaced points across the entire bounding box, such that at least one point will exist for each tile
3. Cut out all the points that lie *outside* a polygon that more tightly covers the congo basin rainforest
4. Use the MGRS library to obtain the tile IDs of each of those points
5. Pull out the unique values from that list and save as a textfile

1 and 2 were already accomplished by Zhenya using the code in the first cell below

```python
#https://stackoverflow.com/questions/20776205/point-in-polygon-with-geojson-in-python

from shapely.geometry import shape, Point
# depending on your version, use: from shapely.geometry import shape, Point

# load GeoJSON file containing sectors
with open('./data/Geometry/congo_basin_boundary/Congo_Basin_Boundary_no_islands_v4.geojson') as f:
    js = json.load(f)
```

```python
js['features']
```

```python
# Need the geometry

js['features'][0]['geometry']
```

```python
# Using the code from the Slack Overflow post

# Put the geometry into a Shapely shape object
polygon = shape(js['features'][0]['geometry'])

lon_lat_2 = []
for coord in lon_lat:
    # Put the coordinates into a shapely Point object
    point = Point(coord[0], coord[1])
    if polygon.contains(point):
        # Append all the coordinates (*not* Shapely point objects)
        # that lie within the polygon
        lon_lat_2.append(coord)
```

```python
import mgrs
```

```python
# Using the code at the bottom of the readme here: https://github.com/hobu/mgrs

m = mgrs.MGRS()
# Test on a single point
c = m.toMGRS(7, 5)
c
```

```python
tile_ids = []

for coord in lon_lat_2:
    # Get its MGRS tile
    c = m.toMGRS(coord[1], coord[0])
    # We only want the first five characters, as that's how
    # the tile ID is saved in the products' metadata
    tile_id = c[0:5]
    tile_ids.append(tile_id)
```

```python
# Put them into a set to only get the unique values

tile_ids_2 = list(set(tile_ids))

len(tile_ids_2)
```

```python
# Write into a textfile

with open('tile_ids_2.txt', 'w') as f:
    for tile_id in tile_ids_2:
        f.write(f'{tile_id}\n')
```
