```python
from datetime import date

api = get_api()

footprint = translate_geojson_to_wkt('./data/republic_of_the_congo_boundary_bounding_box.geojson')

products = api.query(footprint,
                     date=(date(2020,1,1), date(2020,1,31)),
                     area_relation='IsWithin',
                     platformname='Sentinel-2',
                     cloudcoverpercentage=(1,5),
                     raw='2A')

products_df = api.to_dataframe(products)

products_df.head()
```

```python
import pandas as pd


df2 = products_df[~pd.isnull(products_df['tileid'])]

df2.head()
```

```python
df2 = df2[['title', 'tileid']]

df2.head()
```

```python
def find_tile_id(title):
    return title.split('_')[5][1:]


for sentinel_id in df2.index:
    title = df2.loc[sentinel_id, 'title']
    tileid = df2.loc[sentinel_id, 'tileid']
    if find_tile_id(title) != tileid:
        print('Mistake found:', sentinel_id)
```

```python
df3 = products_df.copy()

df3['tileid_new'] = df3['title'].apply(find_tile_id)

df3.head()
```
